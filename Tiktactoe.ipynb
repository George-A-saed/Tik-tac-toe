```python
# class player
class Player:
    def __init__(self):
        self.name = ""
        self.symbol = ""

    def choose_name(self):
        while True:
            name = input("(Letters only):")
            if name.isalpha() == True:
                self.name = name
                break
            print("Invalid name")

    def choose_symbol(self):
        while True:
            symbol = input(f"{self.name} , choose your sybmol(X,O):")
            if symbol.isalpha() and len(symbol) == 1:
                self.symbol = symbol.upper()
                break
            print("Invalid symbol")

```

```python
#class menu
class Menu:
    def display_menu(self):
        print("Welcome to Tik Tak Toe !")
        print("1.Start game")
        print("2.Quit game")
        while True:
            choice = input()
            if choice.isdigit():
                self.choice = choice
                break
            print("Invalid choice")
        return choice
    def display_end(self):
        menu_text = """
        Game Over !
        1.Restart game
        2.Quit game
        """
        print(menu_text)
        while True:
            choice = input()
            if choice.isdigit():
                self.choice = choice
                break
            print("Invalid choice")
        return choice
```

```python
# class board
class Board:
    def __init__(self):
        self.board = [str(i) for i in range (1 , 10)]


    def display_board(self):
        for i in range(0, 9 , 3):
            print("|".join(self.board[i:i + 3]))
            if i < 6:
                print("-" * 5)

    def update_board(self, choice, symbol):
        if self.is_valid_move(choice):
            self.board[choice - 1] = symbol
            return True
        return False

    def is_valid_move(self, choice):
        if self.board[choice - 1].isdigit():
            return True

    def reset(self):
        self.board = [str(i) for i in range (1 , 10)]

```

```python
#class game
class Game:
    def __init__(self):
        self.players = [Player(),Player()]
        self.board = Board()
        self.menu = Menu()
        self.current_player = 0
    def start_game(self):
        choice = self.menu.display_menu()
        if choice == "1":
            self.setup_players()
            self.play_game()
        else:
            self.quit_game()
    def setup_players(self):
        for number ,player in enumerate( self.players , start = 1):
            print(f"Player {number} , Enter your name:")
            player.choose_name()
            player.choose_symbol()
    def play_game(self):
        while True:
            self.play_turn()
            if self.check_win()  or self.check_draw() == True:
                choice =self.menu.display_end()
                if choice =="1":
                    self.restart()
                else:
                    self.quit_game()
                    exit()

    def check_win(self):
        win_com = [
            [3, 4, 5],[0, 1, 2],[6, 7, 8],
            [0, 3, 6],[1 ,4, 7],[2, 5, 8],
            [0, 4, 8],[2, 4 ,6]
        ]

        for combo in win_com:

            if (self.board.board[combo[0]] == self.board.board[combo[1]] == self.board.board[combo[2]]):
                return True
            continue

    def check_draw(self):
       return all (not cell.isdigit() for cell in self.board.board)

    def play_turn(self):
        player = self.players[self.current_player]
        self.board.display_board()
        print(f"Its {player.name}'s turn{player.symbol}")
        while True:
            try:
                cell_choice =int( input("choose from (1 to 9):"))
                if 1 <= cell_choice <= 9 and self.board.update_board(cell_choice , player.symbol):
                    break
                else:
                    print("Invalid movie")
            except ValueError:
                print("Please enter a number between 1 and 9.")

        self.switch_player()
    def switch_player(self):
        self.current_player = 1 - self.current_player
    def quit_game(self):
        print("Thank you for playing")
    def restart(self):
        self.board.reset()
        self.current_player = 0
        self.play_game()

#main
game = Game()
game.start_game()
```
